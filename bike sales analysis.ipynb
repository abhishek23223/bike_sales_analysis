```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
df = df.rename(columns = {
    "avg_daily_distance_(km)" : "avg_daily_distance",
    "price_(inr)" : "price",
    "mileage_(km/l)" : "mileage",
    "year_of_manufacture" : "manufacturing_year",
    "engine_capacity_(cc)" : "engine_cc",
    "resale_price_(inr)" : "resale_price"
})
```

```python
df.columns
```

```python
df["depreciation"] = df["price"] - df["resale_price"]
df["depreciation"] = df["depreciation"].round(2)
```

```python
df.head(5)
```

```python
q1 = df["price"].quantile(0.25)
q3 = df["price"].quantile(0.75)
iqr = q3 - q1
low = q1-(1.5*iqr)
high = q3+(1.5*iqr)

df[(df["price"]>high) | (df["price"]<low)]
```

```python
q1 = df["resale_price"].quantile(0.25)
q3 = df["resale_price"].quantile(0.75)
iqr = q3 - q1
low = q1-(1.5*iqr)
high = q3+(1.5*iqr)

df[(df["resale_price"]>high) | (df["resale_price"]<low)]
```

```python
def segment(price):
    if price < 100000:
        return "Low"
    elif price < 200000:
        return "Medium"
    elif price < 300000:
        return "High"
    else: 
        return "premium"
;
df["price_segment"] = df["price"].apply(segment)
```

```python
df[["price" , "price_segment"]]
```

```python
df["depreciation_percentage"] = (
    (df["price"] - df["resale_price"]) /
     df["price"]
) * 100
```

```python
df["depreciation_percentage"] = df["depreciation_percentage"].round(2)
```
